```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
life_expectancy_file = "lex.csv"
life_expectancy_df = pd.read_csv(life_expectancy_file)

# Calculate average life expectancy per year
df_mean = life_expectancy_df.iloc[:, 1:].mean(numeric_only=True)
years = df_mean.index.astype(int)

# Plot settings
plt.figure(figsize=(10, 6))
plt.plot(df_mean, years, marker='o', linestyle='dashed', color='#A31D1D',
         markerfacecolor='yellow', markersize=8, linewidth=2, label="Life Expectancy")

# Adding a shaded area for better visualization
plt.fill_betweenx(years, df_mean, color='#A31D1D', alpha=0.1)

# Titles and labels with styling
plt.ylabel("Year", fontsize=14, fontweight='bold', color='black')
plt.xlabel("Average Life Expectancy", fontsize=14, fontweight='bold', color='black')
plt.title("Life Expectancy Over Time", fontsize=16, fontweight='bold', color='darkred')

# Grid and legend customization
plt.legend(loc='lower right', fontsize=12, frameon=True, facecolor='white', edgecolor='black')
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.6)

# Display the plot
plt.show()
```
