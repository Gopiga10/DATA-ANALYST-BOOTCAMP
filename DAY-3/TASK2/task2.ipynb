```python
import pandas as pd
import matplotlib.pyplot as plt

income_file = "mincpcap_cppp (1).csv"
income_df = pd.read_csv(income_file)

df_mean = income_df.iloc[:, 1:].mean(numeric_only=True)
years = df_mean.index.astype(int)

plt.figure(figsize=(10, 5))
plt.plot(df_mean, years, marker='o', linestyle='-', color='brown', label="Change in Average Income")
plt.xlabel("Average Income",fontsize=14, fontweight='bold', color='black')
plt.ylabel("Year",fontsize=14, fontweight='bold', color='black')
plt.title("Change in Average Income Over the Years")
plt.legend()
plt.grid(True)
plt.show()
```
