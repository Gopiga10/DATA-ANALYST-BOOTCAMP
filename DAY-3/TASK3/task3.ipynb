```python
import pandas as pd
import matplotlib.pyplot as plt

happiness_file = "hapiscore_whr.csv"
happiness_df = pd.read_csv(happiness_file)

happiness_df.set_index(happiness_df.columns[0], inplace=True)
happiness_df = happiness_df.transpose()

plt.figure(figsize=(12, 6))
for country in happiness_df.columns[:5]:  
    plt.plot(happiness_df[country], happiness_df.index.astype(int), marker='o', linestyle='-', label=country )

plt.xlabel("Happiness Score (WHR)")
plt.ylabel("Year")
plt.title("Happiness Score Trends Over the Years")
plt.legend(title="Countries")
plt.grid(True)
plt.show()
```
