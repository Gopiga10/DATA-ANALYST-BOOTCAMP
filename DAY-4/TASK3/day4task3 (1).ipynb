```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "telco_customer_churn.csv"
df = pd.read_csv(file_path)

# Check and clean missing values
print("Missing Values Before Cleaning:\n", df.isnull().sum())
df_cleaned = df.dropna()
print("\nMissing Values After Cleaning:\n", df_cleaned.isnull().sum())
print("\nDataset Shape After Cleaning:", df_cleaned.shape)

# Set a modern style for better visualization
sns.set_style("whitegrid")

# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("Customer Data Analysis - Telco Churn", fontsize=16, fontweight="bold")

# Gender Distribution
sns.countplot(x='Gender', data=df_cleaned, palette='coolwarm', ax=axes[0, 0])
axes[0, 0].set_title('Customer Distribution by Gender', fontsize=12, fontweight="bold")
axes[0, 0].set_ylabel("Count")
axes[0, 0].grid(axis='y', linestyle='--', alpha=0.7)

# Monthly Charges Distribution
sns.histplot(df_cleaned['Monthly Charge'], bins=30, kde=True, color='purple', ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Monthly Charges', fontsize=12, fontweight="bold")
axes[0, 1].set_xlabel("Monthly Charge")
axes[0, 1].set_ylabel("Count")
axes[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Scatter plot for Total Charges vs Monthly Charge
scatter = sns.scatterplot(x='Monthly Charge', y='Total Charges', data=df_cleaned, hue='Contract', palette='Spectral', ax=axes[1, 0], alpha=0.7)
axes[1, 0].set_title('Total Charges vs Monthly Charge (Segmented by Contract Type)', fontsize=12, fontweight="bold")
axes[1, 0].set_xlabel("Monthly Charge")
axes[1, 0].set_ylabel("Total Charges")
axes[1, 0].grid(True, linestyle='--', alpha=0.5)
axes[1, 0].legend(title='Contract Type')

# Contract Type Distribution
contract_plot = sns.countplot(x='Contract', data=df_cleaned, palette='viridis', ax=axes[1, 1])
axes[1, 1].set_title('Customer Distribution by Contract Type', fontsize=12, fontweight="bold")
axes[1, 1].set_ylabel("Count")
axes[1, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Annotate contract count bars
for p in contract_plot.patches:
    contract_plot.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom', fontsize=10, fontweight="bold")

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "telco_customer_churn.csv"
df = pd.read_csv(file_path)

# Check and clean missing values
print("Missing Values Before Cleaning:\n", df.isnull().sum())
df_cleaned = df.dropna()
print("\nMissing Values After Cleaning:\n", df_cleaned.isnull().sum())
print("\nDataset Shape After Cleaning:", df_cleaned.shape)

# Apply dark background style
plt.style.use("dark_background")
sns.set_palette("coolwarm")

# Create a figure with subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle(" Customer Data Analysis - Telco Churn", fontsize=18, fontweight="bold", color="white")

#  Gender Distribution
sns.countplot(x='Gender', data=df_cleaned, palette='magma', ax=axes[0, 0])
axes[0, 0].set_title(' Customer Distribution by Gender', fontsize=14, fontweight="bold", color="cyan")
axes[0, 0].set_ylabel("Count", fontsize=12, fontweight="bold", color="white")
axes[0, 0].grid(axis='y', linestyle='--', alpha=0.5)

#  Monthly Charges Distribution
sns.histplot(df_cleaned['Monthly Charge'], bins=30, kde=True, color='#FF6B6B', ax=axes[0, 1])
axes[0, 1].set_title(' Distribution of Monthly Charges', fontsize=14, fontweight="bold", color="lightgreen")
axes[0, 1].set_xlabel("Monthly Charge", fontsize=12, fontweight="bold", color="white")
axes[0, 1].set_ylabel("Count", fontsize=12, fontweight="bold", color="white")
axes[0, 1].grid(axis='y', linestyle='--', alpha=0.5)

#  Scatter plot for Total Charges vs Monthly Charge
scatter = sns.scatterplot(x='Monthly Charge', y='Total Charges', data=df_cleaned, hue='Contract', 
                          palette='coolwarm', ax=axes[1, 0], alpha=0.8, edgecolor="white", s=80)
axes[1, 0].set_title(' Total Charges vs Monthly Charge (Segmented by Contract Type)', fontsize=14, fontweight="bold", color="yellow")
axes[1, 0].set_xlabel("Monthly Charge", fontsize=12, fontweight="bold", color="white")
axes[1, 0].set_ylabel("Total Charges", fontsize=12, fontweight="bold", color="white")
axes[1, 0].grid(True, linestyle='--', alpha=0.5)
axes[1, 0].legend(title='Contract Type', fontsize=10, title_fontsize=12, facecolor='black')

#  Contract Type Distribution
contract_plot = sns.countplot(x='Contract', data=df_cleaned, palette='plasma', ax=axes[1, 1])
axes[1, 1].set_title(' Customer Distribution by Contract Type', fontsize=14, fontweight="bold", color="orange")
axes[1, 1].set_ylabel("Count", fontsize=12, fontweight="bold", color="white")
axes[1, 1].grid(axis='y', linestyle='--', alpha=0.5)

#  Annotate bars with counts
for p in contract_plot.patches:
    contract_plot.annotate(f'{p.get_height()}', 
                           (p.get_x() + p.get_width() / 2., p.get_height()), 
                           ha='center', va='bottom', fontsize=12, fontweight="bold", color="white")

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
