```python
print("🚀 Code execution started...")

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Set modern style
plt.style.use("dark_background")
sns.set_palette("coolwarm")

# Load dataset
file_path = "telco_customer_churn.csv"
df = pd.read_csv(file_path)

# Standardize column names
df.columns = df.columns.str.strip().str.lower()

# Convert 'Churn' to numerical values if present
if 'churn' in df.columns:
    df['churn'] = df['churn'].map({'Yes': 1, 'No': 0})

# Convert categorical columns into numerical using one-hot encoding
categorical_cols = df.select_dtypes(include=['object']).columns
df = pd.get_dummies(df, columns=categorical_cols, drop_first=True)

# Select only numerical columns
df_numeric = df.select_dtypes(include=[np.number])

# Compute correlation matrix
corr_matrix = df_numeric.corr()

# Find top features most correlated with churn
if 'churn' in corr_matrix.columns:
    churn_correlation = corr_matrix['churn'].drop('churn').sort_values(ascending=False)
    top_churn_features = churn_correlation.head(5)
    print("\n📊 Top Features Most Correlated with Customer Churn:\n", top_churn_features)

    # 🎯 Visualization: Top Features Affecting Churn
    plt.figure(figsize=(10, 5))
    sns.barplot(x=top_churn_features.index, y=top_churn_features.values, palette="magma")
    plt.title("🔥 Top Features Most Correlated with Customer Churn", fontsize=14, fontweight="bold", color="cyan")
    plt.ylabel("Correlation with Churn", fontsize=12, fontweight="bold", color="white")
    plt.xticks(rotation=30, fontsize=11, fontweight="bold", color="white")
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    
    # Annotate the bars with values
    for index, value in enumerate(top_churn_features.values):
        plt.text(index, value + 0.01, f"{value:.2f}", ha='center', fontsize=12, fontweight="bold", color="white")

    plt.show()

else:
    print("\n❌ ERROR: 'Churn' column not found in correlation matrix.")

# 🎨 Visualization: Heatmap for Feature Correlations
plt.figure(figsize=(12, 6))
sns.heatmap(corr_matrix, annot=False, fmt=".2f", cmap="coolwarm", linewidths=0.5)

plt.title("📌 Feature Correlation Heatmap", fontsize=16, fontweight="bold", color="yellow")
plt.xticks(fontsize=10, fontweight="bold", color="white", rotation=45, ha="right")
plt.yticks(fontsize=10, fontweight="bold", color="white")
plt.show()
```
