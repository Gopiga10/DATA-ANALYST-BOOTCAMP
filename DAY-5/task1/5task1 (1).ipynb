```python

import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS
import numpy as np
```

```python
text = """
Football is more than just a sport; it's a passion that unites millions worldwide.
From breathtaking goals to stunning comebacks, the game never fails to excite.
Teams battle for glory, showcasing skill, teamwork, and determination.
Fans cheer, celebrate, and sometimes cry, creating unforgettable moments.
Tournaments like the World Cup and Champions League make history every year.
Legends are born on the field, inspiring generations to come.
"""
```

```python
custom_stopwords = set(STOPWORDS)
custom_stopwords.update(["game", "play", "teams", "one", "every", "never", "just", "year", "like"])
```

```python
size = 500
x, y = np.ogrid[:size, :size]
mask = (x - size//2) ** 2 + (y - size//2) ** 2 > (size//2 - 10) ** 2  # Circular shape
mask = 255 * mask.astype(int) 
```

```python
wordcloud = WordCloud(
    width=size, height=size,
    background_color="black",  
    stopwords=custom_stopwords,
    mask=mask,  
    contour_color="white",  
    contour_width=2,
    colormap="coolwarm",
    max_words=100,
    random_state=42
).generate(text)
```

```python
fig = plt.figure(figsize=(6, 6), facecolor="black")  # Background for better visualization
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")  # Hide axes
plt.title("Football Passion Word Cloud", fontsize=15, color="white", fontweight="bold")
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import numpy as np
import pandas as pd

# Load the dataset
file_path = "IPL_Match_Highlights_Commentary.csv"
df = pd.read_csv(file_path)

# Combine all commentary text
text = " ".join(df["Commentary"].dropna())

# Define custom stopwords
custom_stopwords = set(STOPWORDS)
custom_stopwords.update(["ball", "over", "run", "runs", "bat", "bowler", "batsman", "player"])

# Create a circular mask
size = 500
x, y = np.ogrid[:size, :size]
mask = (x - size//2) ** 2 + (y - size//2) ** 2 > (size//2 - 10) ** 2
mask = 255 * mask.astype(int)

# Generate the word cloud
wordcloud = WordCloud(
    width=size, height=size,
    background_color="black",
    stopwords=custom_stopwords,
    mask=mask,
    contour_color="white",
    contour_width=2,
    colormap="coolwarm",
    max_words=100,
    random_state=42
).generate(text)

# Plot the word cloud
fig = plt.figure(figsize=(6, 6), facecolor="black")
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("IPL Commentary Word Cloud", fontsize=15, color="white", fontweight="bold")
plt.show()
```
