```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

documents = [
    "Machine learning makes AI powerful and useful.","Deep learning is a subset of machine learning in AI.",
    "Natural Language Processing (NLP) is transforming AI.","AI is shaping the future with automation."
]

count_vectorizer = CountVectorizer()
count_vectors = count_vectorizer.fit_transform(documents)
count_df = pd.DataFrame(count_vectors.toarray(), columns=count_vectorizer.get_feature_names_out())

tfidf_vectorizer = TfidfVectorizer()
tfidf_vectors = tfidf_vectorizer.fit_transform(documents)
tfidf_df = pd.DataFrame(np.round(tfidf_vectors.toarray(), 2), columns=tfidf_vectorizer.get_feature_names_out())

def plot_heatmap(df, title, cmap="magma"):
    plt.figure(figsize=(12, 6))
    sns.heatmap(df, annot=True, cmap=cmap, linewidths=1.5, fmt=".2f", cbar_kws={'shrink': 0.75})
    
    plt.xticks(rotation=45, ha="right", fontsize=12, fontweight="bold", color="darkblue")
    plt.yticks(rotation=0, fontsize=12, fontweight="bold", color="darkred")
    
    plt.title(title, fontsize=16, fontweight="bold", color="darkgreen", pad=15)
    plt.xlabel("Words (Features)", fontsize=14, fontweight="bold", color="brown", labelpad=10)
    plt.ylabel("Documents", fontsize=14, fontweight="bold", color="brown", labelpad=10)

    plt.grid(True, linestyle="--", alpha=0.5)
    plt.show()

plot_heatmap(count_df, " CountVectorizer - Word Frequency Matrix")
plot_heatmap(tfidf_df, " TF-IDF - Word Importance Matrix")

print("\n Extracted Words (Feature Names):")
print(list(count_vectorizer.get_feature_names_out()))

print("\n CountVectorizer Output (Word Frequency Matrix):\n", count_df)

print("\n TF-IDF Output (Word Importance Matrix):\n", tfidf_df)
print("\n Trend & Pattern Analysis:")
print(" AI-related words appear frequently, showing their dominance in the dataset.")
print(" 'Learning' appears multiple times, highlighting its importance in the content.")
print(" TF-IDF assigns lower values to common words and higher values to distinctive words.")
print(" This preprocessing is a crucial step in training AI models like ChatGPT & BERT!")
```
