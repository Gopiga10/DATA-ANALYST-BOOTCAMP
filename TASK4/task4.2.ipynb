```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "cell_phones_total (1).csv"
df = pd.read_csv(file_path)

years = df.columns[1:]  
df_numeric = df[years].apply(pd.to_numeric, errors='coerce')
avg_usage = df_numeric.mean(skipna=True)
plt.figure(figsize=(10, 7))
plt.plot(avg_usage.values, avg_usage.index.astype(int), marker='o', linestyle='-', 
         color='#FF5733', markerfacecolor='yellow', markersize=8, linewidth=2, label="Avg Cell Phone Usage")

plt.fill_betweenx(avg_usage.index.astype(int), avg_usage.values, color='#FF5733', alpha=0.1)
plt.ylabel("Year", fontsize=14, fontweight='bold', color='darkblue')
plt.xlabel("Number of Cell Phones per 100 People", fontsize=14, fontweight='bold', color='darkblue')
plt.title("Rise in Cell Phone Usage Over the Years", fontsize=16, fontweight='bold', color='darkred')
plt.legend(loc='lower right', fontsize=12, frameon=True, facecolor='white', edgecolor='black')
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.6)
plt.show()
```
